```python
# Polynomial Regression Example
"""
This is a Polynomial Regression Example with One Variable

When the polynomial features are added to the regression 
hypothesis, the function will become non-linear

Polynomial Regression is choosen to address non-linearity part
of data to fit the model better

In this example, Polynomial is fit over Designation and Salary variables 

"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Loading the data set
data = pd.read_csv("Emp_Designation_Salary.csv")
print(data.shape)
print(data.head())
```

```python
# Fetching X and Y values
X = data.iloc[:,1:2].values
Y = data.iloc[:,-1].values
```

```python
# Including polynomial function for better fitting of data
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Generates a matrix of consisting of polynomial combinations of features
# less or equal to degree mentioned
poly = PolynomialFeatures(degree=2)
```

```python
# Transformed polynomial X values
poly_features = poly.fit_transform(X)
poly.fit(X,Y)

poly_regression = LinearRegression()

# replacing X values with poly_features
poly_regression.fit(poly_features,Y)

# Normal regression to compare
regressor = LinearRegression()
regressor.fit(X,Y)
```

```python
# Plotting the data for Polynomial features with degree=2
plt.scatter(X,Y)
plt.plot(X,poly_regression.predict(poly_features))
plt.title("Polynomial Regression- Desgination Vs Salary , degree=2 ")
plt.xlabel("Designation")
plt.ylabel("Salary")
plt.show()
```

```python
# Plotting the data for Normal regressor
plt.scatter(X,Y)
plt.plot(X,regressor.predict(X))
plt.title("Linear Regression- Designation Vs Salary ")
plt.xlabel("Designation")
plt.ylabel("Salary")
plt.show()
```

```python
# Including degree=3 to the data for Polynomial features
poly = PolynomialFeatures(degree=3)
poly_features = poly.fit_transform(X)
poly.fit(X,Y)
poly_regression = LinearRegression()
poly_regression.fit(poly_features,Y)
```

```python
# Plotting the data for polynomial degree=3
plt.scatter(X,Y)
plt.plot(X,poly_regression.predict(poly_features))
plt.title("Polynomial Regression- Designation Vs Salary, Degree=3")
plt.xlabel("Designation")
plt.ylabel("Salary")
plt.show()
```
