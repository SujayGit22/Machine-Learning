```python
"""
Iris data set analysis with K-nearest Neighbour model
Libraries used - Scikit-learn, Seaborn,pyPlot,pandas

Iris data has the 3 species of flowers varying in its
Sepal length and width, Petal length and width.

Problem - On unseen data, what is prediction score of the flower to be
either of the species
Solution - K-nearest neighbour is used as the model, to make the prediction.
In this model, the K value is define and based on the voting and euclidean distance 
the new data point is classified to the species.
"""
```

```python
# Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Data loading and exploratory analysis
data = pd.read_csv('D:\Workspace\Machine Learning\KNN\Iris.csv')
print(data.shape)
print(data.head())
print(data.describe())
```

```python
# dropping ID column- not required
data = data.drop(labels=['Id'],axis=1)
```

```python
# Display a grid explaining the relationship between variables;
# Species type define different color codes
sns.pairplot(data, hue='Species')
plt.show()
```

```python
# Split data into training and test data
x_train, x_test, y_train, y_test = train_test_split(data[['SepalWidth','SepalLength',
                                                          'PetalWidth','PetalLength']],
                                                    data['Species'],random_state=0)
# Splitting of Data
print("x_train shape: {}\ny_train shape: {}".format(x_train.shape,y_train.shape))
print("x_test shape: {}\ny_test shape: {}".format(x_test.shape,y_test.shape))
```

```python
# Working and defining the KNN model
# Initializing the model
KNN = KNeighborsClassifier(n_neighbors=1)

# Fit the model
KNN.fit(x_train,y_train)
# Predict on the test data
y_pred = KNN.predict(x_test)
```

```python
# Concatenating Predicted values on the test data
pd.concat([x_test, y_test, pd.Series(y_pred, name='Predicted', index=x_test.index)],
          ignore_index=False, axis=1)
```

```python
# Score test
print("Test set score:{:.2f}".format(KNN.score(x_test,y_test)))
```
